```python
# Импортируем библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Читаем датасет
data = pd.read_csv('/content/Fish.csv', encoding='cp1252')
data.head(10)
```

```python
# Выводим информацию
data.info()
```

```python
# Выводим информацию о типах столбцов
data.dtypes
```

```python
# Названия столбцов и строк
print('Столбцы: ', list(data.columns.values))
print('Строки: ', list(data.index.values))
```

```python
# Количество пустых ячеек
np.count_nonzero(data.isnull().sum(axis= 1).values)
```

```python
# Заменяем категариальные переменные
table = pd.get_dummies(data, columns = ['ï»¿Species'], prefix = '', prefix_sep = '', drop_first = False)
table.head(10)
```

```python
# Матрица корреляции
plt.figure(figsize=(15, 10)).add_subplot(1,1,1)
sns.heatmap(table.corr(), annot = True)
plt.show()
```

```python
# Распределение веса рыб
table['Weight'].plot(kind='hist', bins=20, title='Weight')
```

```python
# Распределение роста рыбы
plt.hist(table['Height'])
```

```python
table.plot(kind='scatter', x='Weight', y='Height')
# Тип ответа не зависит от вида рыбы
```

```python
plt.hist(table['Bream'].astype(int))
# Распределение рыб
# Bream намного меньше
```

```python
plt.hist(table['Pike'].astype(int))
# Распределение рыб
# Pike намного меньше
```

```python
table['Length1'].plot(kind='line', figsize=(8, 4), title='Length1')
# Видим, что длина1 с id от 20 до 40 очень мало
# Больше всего с id от ~5 до ~45
```

```python
table['Length1'].plot(kind='hist', bins=20, title='Length1')
# Интерпритация предыдущего графика в виде гистограммы
```

```python
sns.countplot(x=table["Roach"]);
# Меньшая часть рыб относится к виду Roach
```

```python
data.groupby('ï»¿Species').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
# Распределение студентов виду
```

```python
sns.pairplot(table[['Roach', 'Height']])
# Рыбы с одним из нихких ростов относятся к виду Roach
```

```python
sns.pairplot(table[['Roach', 'Weight']])
# У рыб, относящихся к виду Roach, самый маленький вес
```

```python
sns.pairplot(table[['Length1', 'Length2', 'Length3', 'Width']])
# Зависимости всех столбцов
```
